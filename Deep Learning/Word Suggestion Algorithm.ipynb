## Olah data training

Sebelum data training dimasukkan kedalam algoritma, kita ubah dulu data yang kita punya agar menjadi sekumpulan kalimat dalam sebuah list

```python
kumpulan_kalimat = data.split(", ")

#Kumpulan Kalimat yang digunakan untuk training ada di bawah sini
kumpulan_kalimat
```

Lalu, dari tiap kalimat yang kita punya kita pisahkan menjadi perkata

```python
#Dari tiap kalimat, kita breakdown per kata

kata_perkalimat = []
for kalimat in kumpulan_kalimat:
  kata_perkalimat.append(kalimat.split())

kata_perkalimat
```

## Bangun Algoritmanya

Inti algoritma yang saya bangun disini adalah saya mencari kata dengan frekuensi kemunculan tertinggi pada suku kata ke-N(step) di depan kata yang saya input

```python
#ALGORITMA

def freq_word(input,step=1):
    freq  = {} 
    for kata in kata_perkalimat:
        if kata.count(input): #Evaluasi apakah ada kata yang sama dalam kalimat dengan kata yang kita input
            i = kata.index(input) + step
            try: #Pengunaan try disini berguna agar error yang terjadinya karena index error bisa diskip saja (Ini dilakukan untuk kasus jika didepan kata yang kita input tidak ada kata lain lagi)
                if kata[i] in freq: # Jika kata sudah ada di dalam dictionary freq, maka tinggal update saja frekuensinya (+1)
                    x = freq[kata[i]] 
                    freq[kata[i]] = x + 1
                else: #Jika belum berarti frekuensinya 1
                    freq[kata[i]] = 1
            except IndexError:
                pass

    #Masukkan kata beserta frekuensinya kedalam suatu variabel
    kata = list(freq.keys())  
    freq = list(freq.values())
    n_suggestion = 4 #Ini jumlah kata yang ingin disajikan sebagai suggestionnya

    import heapq #Library ini hanya sekedar untuk mengurutkan element list dari terbesar ke terkecil
    maximum_freq = heapq.nlargest(n_suggestion, freq)

    suggested_words = []

    for max in maximum_freq:
        for n in freq:
            if n == max:
                suggested_words.append(kata[freq.index(n)])
                kata.pop(freq.index(n))
                freq.pop(freq.index(n))

    return suggested_words
```

## Hasil Suggestion

```python
#MASUKKAN KEYWORD YANG INGIN DISUGGEST

keyword = "Cara" #Bisa diubah2 misal: corona, pln, cara, covid-19, vaksin
keyword = keyword.lower()

print("Contoh untuk suggest 1 kata kedepan: ")
for i in freq_word(keyword,1):
    print(keyword,i)

print("\nContoh untuk suggest 2 kata kedepan: ")
for i in freq_word(keyword,1):
    for j in freq_word(keyword,2):
        if j in freq_word(i,1): #Sebisa mungkin kata ke 2 yang diprediksi berdasarkan keyword dan juga kata ke 1 yang diprediksi berdasarkan suggestionnya keyword merupakan kata yang sama agar kalimat yang dibangun masuk akal
            print(keyword,i,j) #Kata yang disuggest berdasarkan 1 kata dan 2 kata didepan digabungkan

print("\nContoh untuk suggest 3 kata kedepan")

for i in freq_word(keyword,1):
    for j in freq_word(keyword,2):
        if j in freq_word(i,1):
            for k in freq_word(keyword,3):
                if k in freq_word(j,1):
                    print(keyword,i,j,k) #Kata yang disuggest berdasarkan 1,2, dan 3 kata didepan digabungkan

```
